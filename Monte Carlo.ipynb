```python
import numpy as np
pi = np.pi
import matplotlib.pyplot as plt
import os
import IPython
#from scipy import stats as stats
from matplotlib import rc
import scipy.io
from scipy.fftpack import dst, idst

rc('text', usetex=True)
plt.rcParams['text.latex.preamble']=[r"\usepackage{amsmath}"]
plt.rcParams['text.latex.preamble'] = [r'\usepackage{amssymb}']
plt.rcParams['text.latex.preamble'] = [r'\boldmath']
plt.rc('font', family='serif', serif='cm10', weight='bold')
#rc('font.weight', medium)
plt.rcParams['axes.linewidth'] = 2 #set the value globally

%matplotlib inline
```

```python
def simulatedx(N, dt, sigma):
    simdxk = sigma * np.random.randn(N)
    temp = np.arange(0,N)
    cols = np.tile(temp, (N,1))
    rows = np.transpose(cols)
    M = np.sin(np.pi*rows*cols/(N+1))*2/(N + 1)/dt
    
    return np.array([np.dot(M,simdxk),simdxk])
    #return np.array([[np.dot(M,simdxk)],[simdxk]])
    
def powx(dx,dt):
    N = dx.size
    dxk = dt/2*dst(dx,1)
    Pxk = 2*dxk**2/(N + 1)/dt
    return Pxk

def boxAv(x, boxSize):
    N = x.size
    numBoxes = np.ceil(N/boxSize)
    lastBox = np.mod(N,boxSize)
    binAv = np.zeros((numBoxes))
    for i in np.arange(0,numBoxes - 1):
        binAv[i] = np.mean(x[i*boxSize:(i+1)*boxSize - 1])

    if lastBox == 1:
        binAv[-1] = x[-1]
    else:
        binAv[-1] = np.mean(x[-1:-(lastBox):-1])
    return binAv

def cveD(dx,dt,R):
    N = dx.size
    dxn_dxn1 = dx*np.roll(dx,-1)
    dxn_dxn1 = dxn_dxn1[0:-1]
    
    D = np.mean(dx**2)/2./dt + np.mean(dxn_dxn1)/dt
    sigmaX2 = R*np.mean(dx**2) + (2*R - 1)*np.mean(dxn_dxn1)
    
    ep = sigmaX2/D/dt - 2*R
    varianceD = D**2*((6 + 4*ep + 2*ep**2)/N + 4*(1+ep)**2/N**2)
    SNR = np.sqrt(D*dt/sigmaX2)
    return np.array([D, sigmaX2, varianceD, SNR])

def theoryPow(D, dt, sigmaX2, N, k):
    return 2.*(1. + (sigmaX2/D/dt - 2./6)*(1. - np.cos(pi*k/(N + 1))))
```

```python
N = 1000
dt = 0.01
[dx, dxk] = simulatedx(N,dt,0.01)
Pxk = powx(dx,0.01)
```

```python
R = 1./6
[D, sigmaX2, varianceD, SNR] = cveD(dx,dt,R)
```

```python
box = 50
k = np.arange(0,N/box)
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.plot(k,boxAv(Pxk/D/dt**2,box),'o')
plt.plot(k,theoryPow(D,dt,sigmaX2, N, k))

plt.subplot(2,2,2)
plt.plot(Pxk/D/dt**2,'o')

plt.subplot(2,2,3)
plt.hist(dx,N/25,facecolor='green')

plt.subplot(2,2,4)
plt.hist(dxk,N/25)
```
